# Importing Data and Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import json
from io import StringIO
pd.options.display.max_columns = 100
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
import azure
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
import os
import re
import yaml
```

```python
# As a class to preserve state
class AzureBlobStorageManager:
    def __init__(self, connection_str: str, container_name: str):
        self.blob_service_client = BlobServiceClient.from_connection_string(connection_str)
        self.container_client = self.blob_service_client.get_container_client(container_name)

    def download_blob(self, blob_name: str):
        """Download a blob from the container """
        # Create a blob client
        blob_client = self.container_client.get_blob_client(blob_name)
        try:
            # Download blob
            blob_data = blob_client.download_blob().readall()
            df = pd.read_csv(StringIO(blob_data.decode('utf-8')))
            print(f"Blob {blob_name} downloaded successfully.")
            return df
        except Exception as e:  # Do something with this exception block (e.g. add logging)
            print(f"An error occurred: {str(e)}")
            logger.error(f"An error occurred: {str(e)}")
            return 0
```

```python
# Testing class and method

connection_string =''    # Azure container/Access Key/Under Key1, Connection String.
container_name = ''  # The name comes from Containers. 

az_blob_manager = AzureBlobStorageManager(connection_str=connection_string,
                                          container_name=container_name)

de_df = az_blob_manager.download_blob('state_data.csv')
```

# System Count Metric (SCM)

```python
scm_hw_model = ExponentialSmoothing(de_df['system_count'],
                                trend='add',
                                seasonal='add',
                                seasonal_periods=12)
scm_hw_model_fit = scm_hw_model.fit()
scm_hw_forecast = scm_hw_model_fit.forecast(steps=12)
```

```python
scm_pred = pd.DataFrame(scm_hw_forecast,columns=['system_count'])
```

```python
print(scm_pred)
```

```python
plt.figure(figsize=(10,3),dpi=120)
de_df['system_count'].plot(label='observed',marker='.')
scm_hw_forecast.plot(marker='.',label='hw forecasted')
plt.legend()
plt.title('System Count Metric for Delaware')
plt.xlabel('Date')
plt.ylabel('System Count');
```

# Total Rebate Metric (TCM)

```python
trm_hw_model = ExponentialSmoothing(de_df['rebate'],
                                trend='add',
                                seasonal='add',
                                seasonal_periods=12)
trm_hw_model_fit = trm_hw_model.fit()
trm_hw_forecast = trm_hw_model_fit.forecast(steps=12)
```

```python
trm_pred = pd.DataFrame(trm_hw_forecast,columns=['rebate'])
```

```python
print(trm_pred)
```

```python
plt.figure(figsize=(10,3),dpi=120)
de_df['rebate'].plot(label='observed',marker='.')
trm_hw_forecast.plot(marker='.',label='hw forecasted')
plt.legend()
plt.title('Total Rebate Metric for Delaware')
plt.xlabel('Date')
plt.ylabel('Total Rebate ($)');
```

# Technology Cost Metric (TCM)

```python
tcm_hw_model = ExponentialSmoothing(de_df['tech_cost($/W)'],
                                trend='add',
                                seasonal='add',
                                seasonal_periods=12)
tcm_hw_model_fit = tcm_hw_model.fit()
tcm_hw_forecast = tcm_hw_model_fit.forecast(steps=12)
```

```python
tcm_pred = pd.DataFrame(tcm_hw_forecast,columns=['tech_cost($/W)'])
```

```python
print(tcm_pred)
```

```python
plt.figure(figsize=(10,3),dpi=120)
de_df['tech_cost($/W)'].plot(label='observed',marker='.')
plt.fill_between(de_df.index, 
                 de_df['tech_cost($/W)_p25'], 
                 de_df['tech_cost($/W)_p75'], 
                 color='grey', alpha=0.15, label='Interquartile Range')
tcm_hw_forecast.plot(marker='.',label='hw forecasted')

plt.legend()
plt.title('Technology Cost Metric for Delaware')
plt.xlabel('Date')
plt.ylabel('Technology Cost ($/W)');
```

# Rebate Efficiency Metric (REM)

```python
rem_hw_model = ExponentialSmoothing(de_df['rebate_eff(W/$)'],
                                trend='add',
                                seasonal='add',
                                seasonal_periods=12)
rem_hw_model_fit = rem_hw_model.fit()
rem_hw_forecast = rem_hw_model_fit.forecast(steps=12)
```

```python
rem_pred = pd.DataFrame(rem_hw_forecast,columns=['rebate_eff(W/$)'])
```

```python
print(rem_pred)
```

```python
plt.figure(figsize=(10,3),dpi=120)
de_df['rebate_eff(W/$)'].plot(label='observed',marker='.')
plt.fill_between(de_df.index, 
                 de_df['rebate_eff(W/$)_p25'], 
                 de_df['rebate_eff(W/$)_p75'], 
                 color='grey', alpha=0.15, label='Interquartile Range')
rem_hw_forecast.plot(marker='.',label='hw forecasted')

plt.legend()
plt.title('Rebate Efficiency Metric for Delaware')
plt.xlabel('Date')
plt.ylabel('Rebate Efficiency (W/$)');
```

# Concatenating Results

```python
historic_data = de_df[['system_count',
                       'rebate',
                       'tech_cost($/W)',
                       'tech_cost($/W)_p25',
                       'tech_cost($/W)_p75',
                       'rebate_eff(W/$)',
                       'rebate_eff(W/$)_p25',
                       'rebate_eff(W/$)_p75']].copy()
```

```python
prediction_data = pd.concat([scm_pred,
                             trm_pred,
                             tcm_pred,
                             rem_pred], axis=1).copy()
```

```python
historic_data.loc[:, 'IsPrediction'] = 0  # For original rows, the indicator is 0
prediction_data.loc[:, 'IsPrediction'] = 1  # For prediction rows, the indicator is 1
```

```python
final_data = pd.concat([historic_data, prediction_data], axis=0)
```

```python
final_data['system_count'] = final_data['system_count'].round(0).astype('int')
```

```python
final_data['rebate'] = final_data['rebate'].round(2)
```

```python
final_data['tech_cost($/W)'] = final_data['tech_cost($/W)'].round(2)
```

```python
final_data['rebate_eff(W/$)'] = final_data['rebate_eff(W/$)'].round(2)
```

```python
final_data.fillna(value=0,inplace=True)
```
